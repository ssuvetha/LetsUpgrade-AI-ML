# mannwhitney test

### h0: there is no significant difference in distance from home between attrition yes and attrition no.
### h1: there is significant difference in distance from home between attrition yes and attrition no.

```python
import pandas as pd
dataset3=pd.read_excel("general_data.xlsx",sheet_name=0)
dataset4=pd.read_excel("general_data.xlsx",sheet_name=1)
from scipy.stats import mannwhitneyu 
yes=dataset3.DistanceFromHome
no=dataset4.DistanceFromHome
stat, p=mannwhitneyu(yes,no)
print(stat,p)
```

 ### as p<0.05 , h1 is accepted

# TTest 

### h0: there is no significant difference in monthly income between attrition yes and attrition no.
### h1: there is significant difference in monthly income between attrition yes and attrition no.

```python
from scipy.stats import ttest_ind
dataset3=pd.read_excel("general_data.xlsx",sheet_name=0)
dataset4=pd.read_excel("general_data.xlsx",sheet_name=1)
yes=dataset3.DistanceFromHome
no=dataset4.DistanceFromHome
stat, p=ttest_ind(no,yes)
print(stat,p)
```

 ### as p<0.05 , h1 is accepted
